## 3. Description of Key Variables

1. **median_income**  
   - Represents the median income of households in a block group (in tens of thousands).
   - **Range**: Approximately 0.5 to 15.
   - **Insight**: Shows a strong positive linear relationship with median house value, making it a key predictor in regression models.

2. **population**  
   - The total number of people living in each geographic block group.
   - **Range**: Varies from under 100 to over 30,000.
   - **Insight**: No strong correlation with house value, suggesting limited usefulness in simple linear regression.

3. **households**  
   - Number of households in each block group.
   - **Range**: From fewer than 50 to over 6,000.
   - **Insight**: Slight positive trend with house value; may be moderately useful when used with other variables.

4. **median_house_value**  
   - The target variable indicating median house value in US dollars.
   - **Range**: Roughly $15,000 to $500,001, with many values capped at $500,001.
   - **Insight**: Skewed distribution due to the ceiling effect; heavily influenced by `median_income`.



### 4. Linear Regression Modeling
improved code with copilot tool


```python
def run_linear_regression(df, feature, target='median_house_value'):
    X = df[[feature]].values
    y = df[target].values

    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)

    mse = mean_squared_error(y, y_pred)
    mae = mean_absolute_error(y, y_pred)

    return {
        'feature': feature,
        'intercept': model.intercept_,
        'slope': model.coef_[0],
        'mse': mse,
        'mae': mae,
        'y_pred': y_pred
    }

results = [
    run_linear_regression(df, 'median_income'),
    run_linear_regression(df, 'population'),
    run_linear_regression(df, 'households')
]
```

### 5. Regression Results Table
improved coding with copilot tool


```python

def train_and_evaluate(feature):
    X = df[[feature]]
    y = df['median_house_value']
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    mse = mean_squared_error(y, y_pred)
    mae = mean_absolute_error(y, y_pred)
    return model.intercept_, model.coef_[0], mse, mae, y_pred

results = {}
for feature in ['median_income', 'population', 'households']:
    intercept, slope, mse, mae, y_pred = train_and_evaluate(feature)
    results[feature] = {
        'intercept': intercept,
        'slope': slope,
        'mse': mse,
        'mae': mae,
        'y_pred': y_pred
    }

pd.DataFrame.from_dict(results, orient='index')
```

## 6. Visualize Regression Lines

improved coding with copilot tool


```python
for feature in ['median_income', 'population', 'households']:
    plt.figure()
    sns.scatterplot(x=df[feature], y=df['median_house_value'], alpha=0.3)
    plt.plot(df[feature], results[feature]['y_pred'], color='red')
    plt.title(f'{feature} vs Median House Value')
    plt.xlabel(feature)
    plt.ylabel('Median House Value')
    plt.text(0.05, 4.5,
             f"Intercept: {results[feature]['intercept']:.2f}\n"
             f"Slope: {results[feature]['slope']:.2f}\n"
             f"MSE: {results[feature]['mse']:.2f}\n"
             f"MAE: {results[feature]['mae']:.2f}",
             bbox=dict(facecolor='white', alpha=0.5))
    plt.show()
```

### 7. Conclusion and Recommendation
Among the three models, the one using median income as the independent variable shows the strongest linear relationship with median house value. It has the lowest Mean Squared Error (MSE) and Mean Absolute Error (MAE), indicating better predictive performance.

The models using population and number of households have weaker linear patterns and higher errors. They are less effective as standalone predictors of house value.

Recommendation: Use median income as the primary feature when building a simple predictive model for median house value. For more complex models, additional variables and non-linear techniques should be explored.
